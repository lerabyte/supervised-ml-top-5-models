# **Block 1 — Imports**

This block brings in the only tools we need: one model (LinearRegression), one function to split data into train/test, and one function to score error (MSE). We also use numpy only to convert our hand-typed lists into the format scikit-learn expects.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **Block 2 — Synthetic dataset (X and y)**

This block defines a tiny fake dataset. X is the input (feature) and y is the number we want to predict (target). In scikit-learn, X must be a 2D list like [[x1], [x2], ...] even if there’s only one feature.

```python
# X = one feature (2D): each row is one datapoint: [x]
X_list = [[1], [2], [3], [4], [5], [6], [7], [8]]

# y = target (1D): one number per row in X
# This follows a simple pattern (roughly y = 3x + 2) with tiny variation
y_list = [5, 8, 11, 14, 17, 20, 23, 26]

# Convert to arrays (just format for sklearn)
X = np.array(X_list)
y = np.array(y_list)
```

# **Block 3 — Train/test split**

This block splits the dataset into training rows (the model learns from these) and test rows (the model never learns from these; we use them to check performance on “new” data).

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,   # 20% test, 80% train
    random_state=1    # same split every time
)
```

# **Block 4 — Train a linear regression model**

This block creates a linear regression model and trains it. Training means learning the best slope and intercept from the training data.

```python
model = LinearRegression()     # make the model
model.fit(X_train, y_train)    # train (learn slope + intercept)

print("slope (m):", model.coef_[0])     # weight for x
print("intercept (b):", model.intercept_)
```

# **Block 5 — Predict and measure error (MSE)**

This block uses the trained model to make predictions on the test set, then computes MSE (mean squared error). Lower MSE means the predictions are closer to the true values on average.

```python
y_pred = model.predict(X_test)                 # predictions (y-hat)
mse = mean_squared_error(y_test, y_pred)       # average squared mistake

print("test MSE:", mse)

# Show a few examples so it feels real
for x_val, true_y, pred_y in zip(X_test.flatten(), y_test, y_pred):
    print("x =", x_val, "| y =", true_y, "| y_hat =", round(pred_y, 2))
```
